```python
# Import library:

import pandas as pd
import numpy as np
from textblob import TextBlob #For processing text, especially natural language. It provides features such as sentiment analysis, part-of-speech tagging, and noun phrase extraction.
import re #A Python library for working with regular expressions, used for pattern matching and text manipulation.
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer  #Used for lexicon-based sentiment analysis. It evaluates the polarity of text (positive, negative, neutral).
from sklearn.model_selection import train_test_split #Splits a dataset into training and testing subsets for machine learning tasks.
from sklearn.metrics import accuracy_score, classification_report #Measures the accuracy of a classification model by comparing predicted and actual labels.
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis #A machine learning algorithm for classification. It reduces dimensionality while preserving as much class separability as possible.
```

# <a id ='2.1'>  2.1. Understand the Dataset: </a>

```python
# Load the data 1:
df1=pd.read_csv('/content/Combined_News_DJIA.csv')
df1.head()
```

```python
#load the data2:

df2= pd.read_csv('/content/upload_DJIA_table.csv')
df2.head()
```

```python
df1.shape, df2.shape
```

```python
# Merge the data st on the date field:

merge= df1.merge(df2, how='inner', on='Date')
merge
```

# <a id ='2.2'>  2.2. Combine news headlines as one block to feed: </a>

```python
#Combine the top news headline:

headline= []

for row  in range (0, len(merge.index)):
    headline.append(" ".join(str(x) for x in merge.iloc[row, 2:27]))
```

```python
# print a sample of combine headlines:

headline[0]
```

# <a id ='2.3'>  2.3. Clean text data: remove breaks and slashes </a>

```python
# clean the data:
clean_headline = []

for i in range (0, len(headline)):
    clean_headline.append(re.sub("b[(')]", '', headline[i])) # remove b'
    clean_headline[i] = re.sub('b[(")]', '', clean_headline[i]) # remove b"
    clean_headline[i] = re.sub("\'", '', clean_headline[i]) # remove \'
```

```python
clean_headline[0]
```

# <a id ='2.4'>  2.4. Data After Clean</a>

```python
# Add the clean headline to the merge data:

merge['Combined_News'] = clean_headline

merge
```

# <a id ='2.5'>  2.5. Get Subjectivity and polarity </a>

```python
# get subjectivity:

def get_subjectivity(text):
    return TextBlob(text).sentiment.subjectivity

# get polarity:

def get_polarity(text):
    return TextBlob(text).sentiment.polarity
```

```python
merge['Subjectivity'] = merge['Combined_News'].apply(get_subjectivity)

merge['Polarity'] = merge['Combined_News'].apply(get_polarity)
```

```python
merge
```

# <a id ='2.6'>  2.6. Sentiment scores using SentimentIntensityAnalyzer </a>

```python
# create function to get the sentiment score:

def getSIA(text):
    sia = SentimentIntensityAnalyzer()
    sentiment= sia.polarity_scores(text)
    return sentiment
```

```python
# get Sentiment score for each day:
compound=[]
neg=[]
pos=[]
neu=[]
SIA=0

for i in range (0, len(merge['Combined_News'])):
    SIA= getSIA(merge['Combined_News'][i])
    compound.append(SIA['compound'])
    neg.append(SIA['neg'])
    pos.append(SIA['pos'])
    neu.append(SIA['neu'])
```

```python
merge['compound']= compound
merge['neg']= neg
merge['pos']=pos
merge['neu']=neu

merge
```

```python
#Create new dataframe contain relevants features:

df= merge[['Label','Open', 'High', 'Low','Volume','Subjectivity','Polarity','compound','neg','pos','neu']]
df
```

# <a id ='2.7'>  2.7. LinearDiscriminantAnalysis Model </a>

```python
#create the feature data set:
X=df

X=np.array(X.drop(['Label'],axis=1))
```

```python
#create the target data set:

y= np.array(df['Label'])
```

```python
# Split the data:

x_train, x_test, y_train, y_test= train_test_split(X,y, test_size=0.2, random_state= 0)
```

```python
#create and train the model:

model= LinearDiscriminantAnalysis().fit(x_train, y_train)

# Model predicton:

predict= model.predict(x_test)

predict
```

# <a id ='2.8'>  2.8. Classification Report </a>

```python
print(classification_report(y_test, predict))
```
